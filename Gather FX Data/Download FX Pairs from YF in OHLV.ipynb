```python
import yfinance as yf
import pandas as pd

pairs = {
    "EURUSD": "EURUSD=X",
    "AUDNZD": "AUDNZD=X",
    "EURGBP": "EURGBP=X",
    "AUDCAD": "AUDCAD=X",
    "CHFJPY": "CHFJPY=X",
}

for name, ticker in pairs.items():
    df = yf.download(ticker, period="5y", interval="1d")
    df.to_csv(f"/Users/joaomartins/Desktop/{name}_daily.csv")
    print(f"Saved {name}: {len(df)} rows")
```
